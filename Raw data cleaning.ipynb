```python
import pandas as pd
```

```python
#loading dataset
data=pd.read_csv(r"C:\Users\clickmoney\Desktop\dataset\raw_data.csv")
data
```

```python
# Check for standard missing values
# Check for non-standard missing values in columns containing strings
non_string_columns = ['Employee Full Name', 'DOB', 'email', 'gender', 'Company', 'Country']

for column in non_string_columns:
    if data[column].dtype == 'object':
        missing_non_standard = data[column].apply(lambda x: x == '$' or x == '').sum()
        print(f"Column '{column}' has {missing_non_standard} non-standard missing values.")
```

```python
data
```

```python
# Fill NaN values with a constant (e.g., 0)
data['Employee Full Name'].fillna(0, inplace=True)
data['DOB'].fillna(0, inplace=True)
data['email'].fillna(0, inplace=True)
data['gender'].fillna(0, inplace=True)
data['Company'].fillna(0, inplace=True)
data['Country'].fillna(0, inplace=True)
data['Income.1'].fillna(0, inplace=True)
data['Income.2'].fillna(0, inplace=True)
data
```

```python
#Handling Income Columns
import numpy as np
data['Income.1'] = data['Income.1'].astype(str).str.replace('$', '').replace('', np.nan).astype(float)
data['Income.2'] = data['Income.2'].astype(str).str.replace('$', '').replace('', np.nan).astype(float)
data
```

```python
#Removing Empty Rows
data.dropna(how='all', inplace=True)  # Remove entirely empty rows
data
```

```python
data.info()
```

```python
data['Income.1']=data['Income.1'].astype(float)
```

```python
# Filling missing values in "Income.1" and "Income.2" columns with their respective means
mean_income_1 = data['Income.1'].mean()
mean_income_2 = data['Income.2'].mean()

data['Income.1'].fillna(mean_income_1, inplace=True)
data['Income.2'].fillna(mean_income_2, inplace=True)
```

```python
# Create a new column "Total Income" by adding "Income.1" and "Income.2" and rounding up
data['Total Income'] = np.ceil(data['Income.1'] + data['Income.2'])

# Display the updated DataFrame
data
```

```python
# Split "Employee Full Name" into "Employee Surname" and "Employee Last Name"
data[['Employee Surname', 'Employee Last Name']] = data['Employee Full Name'].str.split(' ', 1, expand=True)
data
```

```python
# Delete the "Employee Full Name" column
data.drop(columns=['Employee Full Name'], inplace=True)

# Reorder columns to place "Employee Surname" and "Employee Last Name" after "id"
column_order = ['id', 'Employee Surname', 'Employee Last Name'] + [col for col in data.columns if col not in ['id', 'Employee Surname', 'Employee Last Name']]
data = data[column_order]
```

```python
data
```

```python
#replacing values for clear format in dob
data['DOB']= data['DOB'].str.replace("/","-")
data
```

```python
# Group the data by 'Country' and 'Company' and calculate the average total income
average_salary_by_country_company = data.groupby(['Country', 'Company'])['Total Income'].mean().reset_index()

# Print the result
print(average_salary_by_country_company)
```
